```python
label = pd.read_csv('C:\Users\ckraw\Downloads\\labels.txt',sep='\s+',header=None,names=['turk','url','rating'],index_col=False)
gold = pd.read_csv('C:\Users\ckraw\Downloads\\gold.txt',sep='\s+',header=None,names=['url','rating'],index_col=False)
```

```python
gold_url = gold['url'].unique()
mask = label['url'].isin(gold_url)
labels_on_gold = label[mask]
labels_unknown = label[~mask]
labels_unknown
```

```python
Q3.)
```

```python
rater_check = label.merge(gold,left_on='url', right_on='url',suffixes=['_label','gold'])
mask = rater_check['rating_label']==rater_check['ratinggold']
rater_check1=rater_check[mask]
correct = rater_check1['turk'].value_counts()/rater_check['turk'].value_counts()
correct_df = pd.DataFrame(correct,columns=['turk'])
correct_df.reset_index(level=0, inplace=True)
correct_df.columns=['turk','correct']
correct_df
```

```python
t= rater_check[mask]
#t['turk'].value_counts()
tur = pd.DataFrame(t['turk'].value_counts(),columns=['turk'])
tur.reset_index(level=0, inplace=True)
tur.columns=['turk','count']
tur
```

```python
rater_goodness = tur.merge(correct_df,left_on='turk', right_on='turk')
rater_goodness
```

Q7.)

```python
odds = labels_unknown.merge(rater_goodness,left_on='turk', right_on='turk')
odds=odds.groupby(['url','rating'])['odds'].prod()
odds.unstack().fillna(0)
```

```python
rater_count = rater_goodness.sort_values(by='count',ascending=False)
top_gold_turk = rater_count[:len(rater_count)/4]
top_gold_turk['turk']
odds1 = labels_unknown.merge(rater_goodness,left_on='turk', right_on='turk')
odds1 = odds1[['turk','url','rating','odds']].drop_duplicates()
mask5 = odds1['turk'].isin(top_gold_turk['turk'])
odds1=odds1[mask5]
gold_odds=odds1.groupby(['url','rating'])[['odds']].agg(['prod'])
gold_odds
```
